```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
ds=pd.read_csv("/content/insurance_data.csv")
ds.head()
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
#visualize data
plt.scatter(x="age",y="bought_insurance",data=ds)
plt.title("Age vs Bought Insurance")
plt.xlabel('Age')
plt.ylabel('Bought_insurance')
plt.show()
```

```python
#split data
X_train,X_test,y_train,y_test=train_test_split(ds[['age']],ds.bought_insurance,train_size=0.8)
```

```python
#apply algorith to training data
model=LogisticRegression()
#train algorith using data
model.fit(X_train,y_train)
```

```python
#make prediction
y_predicted=model.predict(X_test)
```

```python
#display prediction
y_predicted
```

```python
comparrision=pd.DataFrame({'Actual Values':y_test,'Predicted Values':y_predicted})
comparrision
```
